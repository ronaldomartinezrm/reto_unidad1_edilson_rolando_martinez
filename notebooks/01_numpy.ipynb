```python
import numpy as np  
import pandas as pd
```

```python
#Carga del dataset
df = pd.read_csv("ventas_techventas.csv", encoding='latin-1')

df.head()
```

```python
# Algebra Lineal Aplicada
ingresos = np.dot(df["cantidad"], df["precio_unitario"])
print(ingresos)
ingreso = df["cantidad"] @ df["precio_unitario"]
print(ingreso)
```

```python
# Utilizamos la función np.nan_to_num para convertir los nulos a cero
ingresos = np.dot(np.nan_to_num(df["cantidad"]), np.nan_to_num(df["precio_unitario"]))
print(ingresos)
ingreso = np.nan_to_num(df["cantidad"]) @ np.nan_to_num(df["precio_unitario"])
print(ingreso)
```

```python
# Convertimos la columna precio_unitario a un arreglo estrictamente puro de Numpy
precios = df["precio_unitario"].to_numpy()

# Cálculo de estadísticas
media = np.nanmean(precios)                       # Media
mediana = np.nanmedian(precios)                   # Mediana
desviacion_estandar = np.nanstd(precios)          # Desviación Estándar
percentiles = np.nanpercentile(precios, [25, 75]) # Percentiles
p25, p75 = percentiles[0], percentiles[1]         # Percentelies para imprimir cada uno por separado

# Mostrar resultados formateados
print(f"Media (Promedio): {media:.2f}")
print(f"Mediana: {mediana:.2f}")
print(f"Desviación Estándar: {desviacion_estandar:.2f}")
print(f"Percentil: {percentiles}")
print(f"Percentil 25: {p25:.2f}")
print(f"Percentil 75: {p75:.2f}")
```
